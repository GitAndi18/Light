# Solid angle

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image
```

This chapter deals with the definition and calculation of solid angles.

## Definition

$$
    d \Omega := \frac{\vec{e_R} \, \vec{dA}}{R^2}
$$
$\vec{dA}$ is the surface element considered at distance R.

For an infinitesimally small surface element on a sphere, you get $d \Omega = \text{sin}(\theta)\, d\theta\, d\varphi$. $\theta$ is the angle of latitude and $\varphi$ the angle of longitude.

The following plot shows the sky solid angle with variing $s$, for $a$ = 12 m and $h$ = 12 m (4 storey buildings).

```python
s = np.linspace(0, 12, 30)
a = 12
h = 12
theta = (h - s) / a
Omega = (np.pi - 2 * theta) / np.pi

figure = plt.figure(figsize=(11, 4))
plt.plot(s, Omega)
plt.ylim(0.3, 1.05)
plt.ylabel('$\Omega_{sky}\, /\, \pi$')
plt.xlabel('Height s / m')
plt.grid()
plt.show()
```
